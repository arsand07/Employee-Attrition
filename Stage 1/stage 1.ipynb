### __OUTLIERS__

```python
plt.figure(figsize=(20,15))
for i,j in enumerate(num):
        plt.subplot(3,7,i+1)
        sns.boxplot(y=df[j])
        plt.tight_layout()
```

```python
kolom,out,high,low = [],[],[],[]
for i in num:
        q1 = df[i].quantile(0.25)
        q3 = df[i].quantile(0.75)
        iqr = q3-q1
        lowlimit = q1 - (1.5*iqr)
        highlimit = q3 + (iqr*1.5)

        outlier = df[ (df[i]<lowlimit) | (df[i]>highlimit)].shape[0]

        if outlier>0:
                kolom.append(i)
                out.append(outlier)
                low.append(lowlimit)
                high.append(highlimit)

data = pd.DataFrame({
        'Kolom':kolom,
        'Outliers':out,
        'Low Limit':low,
        'High Limit':high
})

data['Persentase (%)'] = round(data['Outliers']/df.shape[0]*100,2)
data
```

### __KATEGORICAL FEATURES__

Untuk mendapatkan korelasi antar fitur kategorik, kami menggunakan fungsi pearson dengan bantuan library Dython [Sumber disini](https://blog.knoldus.com/how-to-find-correlation-value-of-categorical-variables/)

```python
from scipy.stats import chi2_contingency

p_value=[]
for i in kat:
        data = pd.crosstab(index=df[i], columns=df['Attrition'])        

        chi2 = chi2_contingency(data)
        p_value.append(round(chi2[1],3))

pd.DataFrame({
        'Kolom' : kat.to_list(),
        'P Value':p_value
})
```
